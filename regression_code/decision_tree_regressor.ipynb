```python
#libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
import numpy as np 
from matplotlib import pylab

%matplotlib inline
```

```python
#display the data and data info
df=pd.read_csv('data/USA_Housing.csv')
display(df.info())
display(df.head())
display(df.columns)
display(df.describe())
```

```python
#display the curve of price column
sns.distplot(df[['Price']])
```

```python
X=df[['Area Population','Avg. Area Number of Bedrooms','Avg. Area Number of Rooms','Avg. Area House Age','Avg. Area Income']]
Y=df[['Price']]
```

```python
X = df.iloc[:, 0].values.reshape(-1, 1)  # values converts it into a numpy array
Y = df.iloc[:, 1].values.reshape(-1, 1)
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)

```

```python
X_test = df.iloc[:50, 0].values.reshape(-1, 1)  # values converts it into a numpy array
y_test = df.iloc[:50, 1].values.reshape(-1, 1)  # -1 means that calculate the dimension of rows, but have 1 column

linear_regressor = DecisionTreeRegressor()  # create object for the class
linear_regressor.fit(X_train, y_train)  # perform linear regression
Y_pred = linear_regressor.predict(X_test)  # make predictions
plt.figure(figsize=(20, 5))


plt.scatter(X_test, y_test)
plt.plot(X_test, Y_pred, color='red')
plt.show()
```
